```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#Exercise 8: Random walk with drift

class Coin:
    def __init__(self, bias):
        self.bias = bias
        self.pos = 0
        self.pos_lst = [self.pos]
        
    def toss(self):
        if random.random()<self.bias:
            return True
        else:
            return False
        
    def step_rw(self, steps):
        for i in range(steps):
            if self.toss():
                self.pos += 1
               
            else:
                self.pos -= 1
            self.pos_lst.append(self.pos)
              
```

```python
def simulate(trials, steps):
    a = []
    for i in range(trials):
        coin = Coin(0.6)
        coin.step_rw(steps)
        a.append(coin.pos_lst)
    return a

```

```python
simulations = simulate(200, 30)
```

```python
#random walks
plt.figure(figsize=(12,9))
for i in range(5):
    plt.plot(simulations[i])
    
plt.show()
```
