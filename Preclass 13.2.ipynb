```python
import random
import matplotlib.pyplot as plt
import math
import numpy as np
```

### Exercise 2

```python
class Question4:
    def __init__(self, probability = 0.5, parta = False, partb = False):
        self.position = 0
        self.p = probability
        self.parta = parta
        self.partb = partb
        self.record = [self.position]
        
    def update(self):

        if self.parta == True and self.position == -4:
            self.position = -3
                
        elif self.partb == True and self.position == 6:
            if random.random() < 0.25:
                self.position += 1
            else:
                self.position -= 1
        else:
            if random.random() < self.p:
                self.position += 1
            else:
                self.position -= 1
                
        self.record.append(self.position)

    def display(self, b=20):
        
        plt.hist(self.record, bins = b)
     
```

```python
#The walk without any rules. Purely random

sim = Question4()

for i in range(10000):
    sim.update()
    
sim.display()
```

```python
#When the rule in part a is implemented

sim = Question4(parta = True)

for i in range(10000):
    sim.update()
    
sim.display()
```

```python
#Part b rule is implemented

sim = Question4(partb = True)

for i in range(10000):
    sim.update()
    
sim.display()
```

```python
#Both a and b rules are turned on

sim = Question4(parta = True, partb = True)

for i in range(10000):
    sim.update()
    
sim.display()
```

### Exercise 4

```python
class Question4:
    def __init__(self, var):
        self.x = 0
        self.y = 0
        self.recordx = [self.x]
        self.recordy = [self.y]
        self.std = np.sqrt(var)
        
    def update(self):
        distance = np.random.normal(0, self.std)
        degree = np.random.randint(0, 360)
        
        self.x, self.y = math.cos(math.radians(degree))*distance, math.sin(math.radians(degree))*distance
        
        self.recordx.append(self.x)
        self.recordy.append(self.y)
```

```python
sim1 = Question4(0.5)
sim2 = Question4(1)
sim3 = Question4(2)
```

```python
for i in range(1600):
    sim1.update()
    sim2.update()
    sim3.update()
    
```

```python
#Variance = 0.5

plt.hist2d(sim1.recordx, sim1.recordy, bins=40, cmap=plt.cm.BuGn_r)
```

```python
#Variance = 1

plt.hist2d(sim2.recordx, sim2.recordy, bins=40, cmap=plt.cm.BuGn_r)
```

```python
#Variance = 2

plt.hist2d(sim3.recordx, sim3.recordy, bins=40, cmap=plt.cm.BuGn_r)
```

### Exercise 9

```python
class Question9:
    
    def __init__(self):
        self.x = 0
        self.y = 0
        self.hole = [(13, 7), (13, 8), (14, 7), (14, 8)]
        
    def update(self):
        if (self.x, self.y) in self.hole:
            pass
        else:
            if random.random()<0.5:
                if random.random()<0.5:
                    self.x += 1
                else:
                    self.x -= 1
            else:
                if random.random()<0.5:
                    self.y += 1
                else:
                    self.y -= 1
                    
        if self.x > 20:
            self.x = 20
        if self.x < -20:
            self.x = -20
        if self.y > 20:
            self.y = 20
        if self.y < -20:
            self.y = -20
```

```python
#After 20 steps

finalx = []
finaly = []

for i in range(200):
    sim = Question9()
    
    for j in range(20):
        sim.update()
        
    finalx.append(sim.x)
    finaly.append(sim.y)
    
plt.hist2d(finalx, finaly, bins=40, cmap=plt.cm.BuGn_r)
```
