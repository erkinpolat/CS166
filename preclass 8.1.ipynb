```python
from scipy.optimize import root
import numpy as np

def calculate_q(k):
    '''
    Use a numerical root finder to determine q from the equation
    q = exp(k*(q-1)).
    '''
    return root(lambda q: q - np.exp(k * (q - 1)), 0).x[0]
```

```python
qs = []

for i in range(1, 11):
    qs.append(calculate_q(i))
```

```python
from pylab import *
import networkx as nx

p = 0.0001
pdata = []
gdata = []

while p < 0.1:
    pdata.append(p)
    g = nx.erdos_renyi_graph(100, p)
    ccs = nx.connected_components(g) 
    gdata.append(max(len(cc) for cc in ccs))
    p *= 1.1


loglog(pdata, gdata)
xlabel('p')
ylabel('size of largest connected component')
show()
```
